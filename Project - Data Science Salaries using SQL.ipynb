```python
# import sqlalchemy and create a sqlite engine
from sqlalchemy import create_engine
engine = create_engine('sqlite://', echo=False)

# export the dataframe as a table 'playstore' to the sqlite engine
df.to_sql("df", con =engine)
```

```python
## Test the code by select the complete table
# Obtain the first 5 rows

sql='''

Select * from df
limit 5


''';

    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head()
```

```python
# Obtain the total number of records from this table
sql='''

SELECT COUNT(*) AS [Total Number of Records]
FROM df;


''';

    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql
```

```python
# Check whether the attribute "work_year" has any missing value

sql='''



SELECT COUNT(*) AS [Missing Value Count]
FROM df
WHERE work_year IS NULL;




''';

    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql
```
